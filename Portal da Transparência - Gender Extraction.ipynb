### Coletar nomes masculinos 

```python
# import package
from requests import get

# Check for one single name, known to exist
url_name = 'https://www.dicionariodenomesproprios.com.br/abner/'
response = get(url_name)

# extract the text
text_names = response.text

print(text_names[:500])
```

```python
from bs4 import BeautifulSoup

html_name_soup = BeautifulSoup(response.text, 'html.parser')
type(html_name_soup)
```

```python
gender_value = html_name_soup.select('p.origem-related strong')[0].text
print(gender_value)
```
