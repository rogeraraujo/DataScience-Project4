```python
# import package
from requests import get

# specify the url
url = 'http://www.portaltransparencia.gov.br/servidores/OrgaoExercicio-ListaOrgaos.asp?CodOS=15000&Pagina=5'

# packages the request, send the request and catch the response
response_1 = get(url)

# extract the text
text = response_1.text

print(text[:500])
```

```python
from bs4 import BeautifulSoup

html_soup_1 = BeautifulSoup(response_1.text, 'html.parser')
type(html_soup)
```

```python
#table_rows_1 = html_soup.find_all('tr')
table_rows_1 = html_soup_1.select('div #listagem tr')
print(type(table_rows_1))
print(len(table_rows_1))
```

```python
institution_name = 'UNIVERSIDADE FED. DO RIO GRANDE DO NORTE'
employee_count = 0

for row in table_rows_1:
    columns = row.find_all('td')
    if (columns is None) or (len(columns) < 3):
        continue
    #print(columns[1].text)
    if columns[1].text.strip() != institution_name:
        continue
    employee_count = int(columns[2].text)
    
print(employee_count)
```

```python
# Redeclaring the lists to store data in
full_names = []

# Preparing the monitoring of the loop
start_time = time()
requests = 0

# For every page in the interval 1-413
for page in pages:
    #url 
    url = 'http://www.portaltransparencia.gov.br/servidores/' + \
          'OrgaoExercicio-ListaServidores.asp?CodOrg=26243&' + \
          'Pagina={}'.format(page)
        
    # Make a get request
    response = get(url)
        
    # Pause the loop
    sleep(randint(8,15))
        
    # Monitor the requests
    requests += 1
    elapsed_time = time() - start_time
    print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    clear_output(wait = True)
              
    # Throw a warning for non-200 status codes
    if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(requests, response.status_code))
              
    # Break the loop if the number of requests is greater than expected
    if requests > 413:
        warn('Number of requests was greater than expected.')  
        break 
        
    # Parse the content of the request with BeautifulSoup
    page_html = BeautifulSoup(response.text, 'html.parser')
        
    # Select all the 16 employee rows from a single page
    table_rows = page_html.select('div #listagem tr')
        
    # For every row of these 16, starting from the second
    for row in table_rows[1:]:
        # Scrape the full name
        full_name = row.select('td a')[0].text
        full_names.append(full_name.strip())
```

```python
employee_names = pd.DataFrame({'full_name': full_names})
print(employee_names.info())
employee_names.head(100)
```

```python
employee_names.to_csv('employee_names.csv')
print('Done!')
```
