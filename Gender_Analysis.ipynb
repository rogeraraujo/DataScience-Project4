```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
employee_gender = pd.read_csv("employee_names-7.csv")
```

```python
employee_gender.head()
```

```python
employee_gender.gender.replace(['masculino', 'feminino'], ['male', 'female'], inplace=True)
employee_gender.head()
```

```python
#Remover colunas
employee_gender.drop('Unnamed: 0', axis=1, inplace=True)
employee_gender.drop('Unnamed: 0.1', axis=1, inplace=True)
employee_gender.drop('Unnamed: 0.1.1', axis=1, inplace=True)
employee_gender.columns
```

```python
employee_gender.to_csv('employee_names-8.csv')
```

```python
total_masc = len(employee_gender[employee_gender['gender'] == 'male'])
total_masc
```

```python
total_fem = len(employee_gender[employee_gender['gender'] == 'female'])
total_fem
```

```python
percent = [total_masc, total_fem]
sex = ['male', 'female']
```

```python
employee_gender.values
```

```python
employee_gender.gender.value_counts().plot(kind='bar')
plt.title('Gender of all UFRN employees')
plt.ylabel('Number of Employees')
plt.xlabel('Gender')
plt.show()
```

```python
employee_gender.gender.value_counts().plot(kind='pie', labels=sex, autopct='%1.2f%%')
plt.title('Gender of all UFRN employees')
plt.axis('equal')
plt.show()
```
