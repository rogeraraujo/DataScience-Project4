```python
import pandas as pd

employee_names = pd.read_csv('gender.csv')
employee_names.drop('Index1', axis=1, inplace=True)
employee_names.drop('Index2', axis=1, inplace=True)
employee_names.columns
```

```python
print(employee_names.info())
employee_names.head(100)
```

```python
unique_first_names = employee_names['first_name'].unique()
print(len(unique_first_names))
```

```python
unknown_unique_first_names = [ ]

for i, row in employee_names.iterrows():
    if row['gender'] == 'unknown':
        first_name = row['first_name']
        if first_name not in unknown_unique_first_names:
            unknown_unique_first_names.append(first_name)

print(len(unknown_unique_first_names))
```

### Check gender for unknown names

```python
import rpy2.interactive as r
import rpy2.interactive.packages
r.packages.importr("genderBR")
rlib = r.packages.packages
rlib.genderBR
```

```python
unknown_first_name_df = pd.DataFrame({'first_name': unknown_unique_first_names})
unknown_first_name_df['gender'] = unknown_first_name_df.apply(lambda row: 'unknown', axis=1)
print(unknown_first_name_df.info())
unknown_first_name_df
```

```python
from time import sleep
from random import randint
```

```python
from time import time
from IPython.core.display import clear_output

start_time = time()
requests = 0

for _ in range(5):
    # A request would go here
    requests += 1
    sleep(randint(1,3))
    current_time = time()
    elapsed_time = current_time - start_time
    print('Request: {}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    clear_output(wait = True)
```

```python
# Preparing the monitoring of the loop
start_time = time()
requests = 0

# For all first names whose gender is unknown
for i, row in unknown_first_name_df.iterrows():
    # Loop control
    #if i > 3:
    #    continue

    # Checks the gender
    first_name = row['first_name']
    gender = row['gender']

    try:
        resp_array = rlib.genderBR.get_gender(first_name)
        if (resp_array is not None) and (len(resp_array) > 0):
            gender = resp_array[0]
    except:
        gender = ''

    # Pause the loop
    sleep(randint(5,10))
        
    # Monitor the requests
    requests += 1
    elapsed_time = time() - start_time
    print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    clear_output(wait = True)

    # Postprocessing
    try:
        gender = gender.lower()
    except:
        gender = ''

    if (gender not in ['male', 'female']):
        gender = 'unknown'
    
    row['gender'] = gender

    #print(first_name + ' / ' + gender)
```

```python
unknown_first_name_df.to_csv('unknown_first_names.csv')
print('Done!')
```

```python
for i, row in employee_names.iterrows():
    if row['gender'] != 'unknown':
        continue
    
    first_name = row['first_name']

    rows_ufndf = unknown_first_name_df.loc[unknown_first_name_df['first_name'] == first_name]

    if rows_ufndf.shape[0] < 1:
        continue
    
    #print(rows_ufndf.iloc[0])
    #print(rows_ufndf.iloc[0]['first_name'])
    #print(first_name)
    #print('-')
    
    correct_gender = rows_ufndf.iloc[0]['gender']
    
    if (correct_gender != 'unknown'):
        row['gender'] = correct_gender

print('Done!')
```
